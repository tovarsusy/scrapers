## Twitter Scrapper Cell
**TAKES SEVERAL DAYS TO RUN**

This cell is my actual twitter scrapper. For this project I choose to focus on the months of **February** and **March**. This cell generates a csv per city that gets put into a folder called **data**. The scrapper does a search for each of the key words on a 7 day window. I choose to focus on these two months becasue February is the month people started hearing about the coroanvirus and March is when people started taking the virus more seriously. March is also when a lot disaster declarations were put in place by counties as well as shelter in place and stay at home orders. 

I choose to only use a subset of the cities in Texas because there are so many and I did not have to the time to scrape all the cities in Texas. The subset of cities I used have a population greater than 10,000.

Some cities have to be run seperatly using TX instead of Texas.


```python
#create a function to extract twitter information into a pandas df
def get_twitter_info():
    tweet_df["tweet_text"] = tweet_df["got_criteria"].apply(lambda x: x.text)
    tweet_df["date"] = tweet_df["got_criteria"].apply(lambda x: x.date)
    tweet_df["hashtags"] = tweet_df["got_criteria"].apply(lambda x: x.hashtags)
    tweet_df["link"] = tweet_df["got_criteria"].apply(lambda x: x.permalink)

# set search variables    
keywords= ["#coronavirus",
           "#covid19",
           '#CoronavirusPandemic',
           '#CoronaOutbreak',
           '#covid_19',
           'coronavirus',
           '#coronavirusupdate',
           '#coronavirusoutbreak',
           '#coronavid19',
           'covid-19',
           '#COVID19Pandemic',
           '#quarantine',
           '#CoronaLockdown',
           '#StayAtHome',
           'shelter in place',
           '#shelterinplace',
           '#toiletpapercrisis',
           '#ToiletPaperApocalypse',
           ]


DIR = 'data'
current_city = (len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR,name))]))
# setting city starting point for scrapper
cities_df=pd.read_csv('texas_cities_trimmed.csv')
cities=cities_df.city[current_city:]

#create twitter info for each city
for city in cities:
    city = '{}, Texas'.format(city)
    city_tweets = []
    while start_date < final_date:
        for i, word in enumerate(keywords):
            tweetCriteria = got.manager.TweetCriteria().setQuerySearch(word)\
                                                       .setSince(str(start_date))\
                                                       .setUntil(str(end_date))\
                                                       .setNear(city)
            while True:
                try:
                    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
                    city_tweets += tweets
                    break
                except Exception as e:
                    print('Failed... Trying again!')
                    print(e)
                    time.sleep(60)
            print('Finished keyword: ', keywords[i])
            #time.sleep(60)
            
        print('Finished dates: ', str(start_date), '-', str(end_date))
        start_date += time_delta
        end_date += time_delta
        if end_date > final_date:
            end_date = final_date
    city_tweets = list(set(city_tweets))
    print('Finished: ', city)
    
    
    start_date = datetime.fromisoformat('2020-02-01').date()
    end_date = datetime.fromisoformat('2020-02-07').date()
    
    # creating city csv's
    df = pd.DataFrame(
        [(tweet.text, tweet.date, tweet.hashtags, tweet.permalink) for tweet in city_tweets]
    )
    df.columns = ['tweet_text', 'date', 'hashtags', 'permalink']
    df.to_csv('data/{}.csv'.format(city.split(',')[0]), index=False)
```

## Mapping Cities to Counties
My model will be at a county level not a city level so I need to map cities to counties.

```python
# reading in files
mypath='data'
csvs = [f for f in listdir(mypath) if isfile(join(mypath, f))]
mypath+='/{}'

counties=pd.read_csv('texas_cities_trimmed.csv')

# adding in city and county column
for csv in csvs:
    csv_input = pd.read_csv(mypath.format(csv))
    city = csv.split('.')[0]
    csv_input['city']= city

    csv_input['counties'] = counties.county_name_all[counties.city == city].iloc[0]
    csv_input.to_csv(mypath.format(csv), index=False)
```
